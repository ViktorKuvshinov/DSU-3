Пользователь вводит два числа. Одно присваивается одной переменной, а второе – другой. Необходимо поменять значения переменных так, чтобы значение первой оказалось во второй, а второй – в первой.

```python
num1 = int(input('Число 1: '))
num2 = int(input('Число 2: '))
summ = num2 + num1
print(f' Число 1: {summ - num1}, Число 2: {summ - num2}')
```

Вводятся три целых числа. Определить какое из них наибольшее.

```python
num_1 = int(input('Число 1: '))
num_2 = int(input('Число 2: '))
num_3 = int(input('Число 3: '))

if num_1 > num_2 and num_1 > num_3:
    print(num_1)
elif num_2 > num_1 and num_2 > num_3:
    print(num_2)
else:
    print(num_3)
```

Дана следующая функция y=f(x):
y = 2x – 10, если x > 0
y = 0, если x = 0
y = 2 * |x| – 1, если x < 0

Требуется найти значение функции по переданному x.

```python
var_x = int(input('Введите Х: '))
if var_x > 0:
    res = var_x * 2 - 10
    print(f' Значение y = {res}')
elif var_x < 0:
    res = 2 * abs(var_x) - 1
    print(f' Значение y = {res}')
elif res == 0: 
    print(f' Значение y = 0')
```

Вводятся два целых числа не равных нулю. Проверить делится ли первое на второе. Вывести на экран сообщение об этом, а также остаток (если он есть) и частное (в любом случае).

```python
num_11 = int(input('Число 1: '))
num_22 = int(input('Число 2: '))
if num_11 % num_22 == 0:
    print(f' Делится нацело, ответ {num_11 // num_22}')
else:
    print(f' Не деллится нацело, ответ {int(num_11 / num_22)} целых, остаток: {num_11 % num_22}')
```

Перевести число, введенное пользователем, в байты или килобайты в зависимости от его выбора.

```python
num33 = int(input())
var = input('Для перевода из кб в б, наберите b, для обратного k')
if var == 'b':
    print(f' в {num33} кбайтах - {num33 * 1024} байтов')
else:
    print(f' в {num33} байтах - {num33 / 1024} kбайтов')
```

Найти сумму и произведение цифр, введенного натурального числа. Например, если введено число 325, то сумма его цифр равна 10 (3+2+5), а произведение 30 (3*2*5).

```python
num44 = int(input())
summa = 0
numm_pr = 1
while num44 > 0:
    summa += num44 % 10
    numm_pr *= num44 % 10
    num44 = num44 // 10
print('сумма', summa)
print ('произведение', numm_pr)
```

Вывести на экран ряд натуральных чисел от минимума до максимума с шагом. Например, если минимум 10, максимум 35, шаг 5, то вывод должен быть таким: 10 15 20 25 30 35. Минимум, максимум и шаг указываются пользователем (считываются с клавиатуры).

```python
num_min = int(input('минимальное число:'))
num_max = int(input('максимальное число:'))
num_step = int(input('шаг вывода:'))
for i in range(num_min, num_max, num_step):
    print(i)
```

Вычислить факториал введенного числа.

```python
num55 = int(input('Введите число: '))
fact = 1
for i in range(1, num55 + 1):
    fact *= i
print(fact)
```

Вывести таблицу значений функции y = -0.23x2 + x. Значения аргумента (x) задаются минимумом, максимумом и шагом. Например, если минимум задан как 1, максимум равен 3, а шаг 0.5. То надо вывести на экран изменение x от 1 до 3 с шагом 0.5 (1, 1.5, 2, 2.5, 3) и значения функции (y) при каждом значении x. 

```python
z_xmin = float(input('минимальное значение Х = '))
z_xmax = float(input('максимальное значение Х = '))
x_step = float(input('шаг Х = '))
while z_xmin < z_xmax:
    res_y = round((-0.23 * z_xmin ** 2 + z_xmin), 2)
    print(res_y)
    z_xmin += x_step
```
